# Check the presence of outliers in the dataset, use appropriate plots for visualization, if outliers present? Treat them accordingly.

```python
import seaborn as sns
```

```python


# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\Admin\Downloads\emp_attrition.csv")

# Visualize outliers using box plots
sns.boxplot(data=df)


```

```python
# Detect and treat outliers using Z-score method
# Calculate Z-scores for each column
z_scores = np.abs((df - df.mean()) / df.std())

```

```python
# Set the threshold for Z-score, any value above this threshold will be considered an outlier
threshold = 3

```

```python
# Identify outliers by checking if Z-score exceeds the threshold
outliers = (z_scores > threshold).any(axis=1)

```

```python
# Treat outliers by replacing them with NaN or other appropriate values
df[outliers] = np.nan

```

```python
# Alternatively, you can drop rows with outliers using:
# df = df[~outliers]

# Display the updated DataF
df
```

### Identify the top 3 departments with the highest attrition rate along with their percentage and visualize using plot

```python
# Calculate the attrition rate for each department
attrition_by_dept = df.groupby('Department')['Attrition'].value_counts(normalize=True).rename('percentage').mul(100).reset_index()


```

```python
# Identify the top 3 departments with the highest attrition rate
top_3_depts = attrition_by_dept.loc[attrition_by_dept['Attrition'] == 'Yes'].sort_values(by='percentage', ascending=False).head(3)
```

```python
top_3_depts
```

```python
# Visualize using plot
import matplotlib.pyplot as plt
plt.bar(top_3_depts['Department'], top_3_depts['percentage'])
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.title('Top 3 Departments with Highest Attrition Rate')
plt.show()
```

### Create a bar chart to visualize the distribution of attrition rates by Gender

```python

import matplotlib.pyplot as plt



# Convert 'Attrition' column to integers (1 for attrition, 0 for no attrition)
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Calculate the attrition rate for each gender
attrition_rate = df.groupby('Gender')['Attrition'].mean() * 100

# Reset the index of the attrition_rate series
attrition_rate = attrition_rate.reset_index()

# Plot the bar chart
plt.bar(attrition_rate['Gender'], attrition_rate['Attrition'])
plt.xlabel('Gender')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Gender')
plt.show()

```
